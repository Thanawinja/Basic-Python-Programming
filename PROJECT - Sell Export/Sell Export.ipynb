Prepare Data

```python
# Delete Rows with a Missing value in the column Country

data = data[data["Country"].notnull()]
```

```python
# Rename Columns

data = data.rename(columns = {"Item Type" : "Item_Type",
                              "Sales Channel" : "Sales_Channel",
                              "Order Priority" : "Order_Priority",
                             "Order Date" : "Order_Date",
                              "Order ID" : "Order_ID",
                              "Ship Date" : "Ship_Date",
                              "Units Sold" : "Units_Sold",
                              "Unit Price" : "Unit_Price",
                              "Unit Cost" : "Unit_Cost",
                              "Total Revenue" : "Total_Revenue",
                              "Total Cost" : "Total_Cost",
                              "Total Profit" : "Total_Profit"
                             })
```

```python
list(data.columns)
```

```python
# Create new column
# Add Column %Margin 

data["%Margin"] = ((data["Total_Profit"] / data["Total_Revenue"])*100).round(2)
```

```python
data.head()
```

```python
# Create a function -> Are sales good or bad ?

def if_condition(row):
    if row["%Margin"] > 30:
        val = "good"
    else:
        val = "bad"
    return val
```

```python
# Create new column
# Add Column Dicision 

data["Dicision"] = data.apply(if_condition, axis = 1) # axis = 1 -> columns 
```

```python
# Manipulate Data
# Total_Revenue = Unit_Price x Units_Sold 
# Total_Cost = Unit_Cost x Units_Sold
# Total_Profit = Total_Revenue - Total_Cost
# %Margin = Total_Profit / Total_Revenue

data["Total_Revenue"] = (data["Units_Sold"] * data["Unit_Price"]).round(2) 
data["Total_Cost"] = (data["Units_Sold"] * data["Unit_Cost"]).round(2)
data["Total_Profit"] = (data["Total_Revenue"] - data["Total_Cost"]).round(2)
data["%Margin"] = ((data["Total_Profit"] / data["Total_Revenue"]) *100).round(2)
```

Machine Learning For bussiness

```python
# Import library 

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Check Version

print(f'sklern version : {sklearn.__version__}')
```

```python
# Define a Model (Regression)
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

model = make_pipeline(StandardScaler(with_mean=False), LinearRegression(n_jobs = -1))

model
```

```python
# x is dependent variables and  y is independaent variables 

x = data[['Total_Revenue','Total_Cost']]
y = data['%Margin'] 
```

```python
model.fit(x,y)
```

```python
x
```

```python
# R-square

model.score(x,y)
```

```python
# Slope

model.coef_ 
```

```python
# Intercept

model.intercept_
```

```python
# If Total_Revenue = 50,000 and Total_Cost = 20,000
# Predict %Margin 

model.intercept_ + (-3.38013701e-05* 50000)+(-4.75858163e-05*20000)
```

```python
test_x = pd.DataFrame({'Total_Revenue': [50000], 'Total_Cost': [20000]})
test_x
```

```python
# We can use .predict to predict the model -> Easier

model.predict(test_x)
```

```python
test_2x = pd.DataFrame({'Total_Revenue': [50000, 70000], "Total_Cost" : [20000, 60000]})
test_2x
```

```python
# If we want to predict 2 variables 

model.predict(test_2x)
```

```python
# Let's create a y_predict for every data

y_predict = model.predict(data[["Total_Revenue","Total_Cost"]]) 
y_predict
```

```python
# Create new columns which is the actual value to compare with the model we created

y_actual = data[["%Margin"]]
```

```python
# Mean Square Error 

mean_squared_error(y_actual, y_predict)
```

```python
# Root Mean Square Error 
# Error +- 9.00 , Ex; Actual = 30  Model = 30 +- 9

mean_squared_error(y_actual, y_predict) ** 0.5  
```

```python
# Create new column
# Add Column Model 

data["Predict_%Margin"] = model.predict(data[["Total_Revenue","Total_Cost"]])
data.drop(["Predict_%Margin"], axis = 1)
data.head()
```

```python
# Take 1000 samples of the data in Descending order

margin_top1000_model = data.sort_values("%Margin", ascending = False).head(1000)
```

```python
# Plot Model vs Actual , Paster is Actual and Muted is Prediction 

f, ax = plt.subplots(figsize=(6, 15))

# Actual

sns.set_color_codes("pastel")
sns.barplot(x = "Year", y = "%Margin", data = margin_top1000_model,
            label = "Actual", color = "r")

# Predict 

sns.set_color_codes("muted")
sns.barplot(x = "Year", y="Predict_%Margin", data = margin_top1000_model,
            label = "Prediction", color = "b")

# Add a legend and informative axis label

ax.set(ylabel = "",
       xlabel = "Actual Vs Prediction")
sns.despine(left = True, bottom = True)
```
